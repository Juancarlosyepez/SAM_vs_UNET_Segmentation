# Segementacion de mascaras con modelo SAM con la libreria especializada en imagenes satelitales Samgeo

## Carga de datos de test almacenada con el notebook de segmentacion unet

```python
import numpy as np

# Cargar el array desde el archivo x_test.npy
X_test = np.load('x_test.npy')

# Verificar el contenido del array
print(X_test[0])
```

# Implementacion del modelo SAM con la libreria samgeo

```python
#from tensorflow.keras.models import load_model
import tensorflow as tf
from samgeo import SamGeo
import os
# Initialize SAM class
sam = SamGeo(
    model_type="vit_h",
    checkpoint="sam_vit_h_4b8939.pth",
    sam_kwargs=None,
)

# Segment the test images and derive masks
test_images_dir = 'test_images'
os.makedirs(test_images_dir, exist_ok=True)

segmented_masks = []
for i in range(len(X_test)):
    image_path = os.path.join(test_images_dir, f'test_image_{i}.tif')
    mask_path = os.path.join(test_images_dir, f'test_mask_{i}.tif')

    # Save the test image
    tf.keras.preprocessing.image.save_img(image_path, X_test[i])

    # Segment the image
    sam.generate(image_path, mask_path, batch=True, foreground=True, erosion_kernel=(3, 3), mask_multiplier=255)

    # Load the generated mask
    segmented_mask = tf.keras.preprocessing.image.load_img(mask_path, color_mode='rgb')
    segmented_mask = tf.keras.preprocessing.image.img_to_array(segmented_mask).astype(np.float32) / 255.0
    segmented_masks.append(segmented_mask)

segmented_masks = np.array(segmented_masks)
```

# Visualizacion aleatoria de mascara segmentada por el modelo SAM

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(X_test[i])
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(segmented_mask)
plt.title('Máscara Segmentada')

plt.show()
```

# Guardar mascaras segmentadas por SAM

```python
# Suponiendo que segmented_masks es un array de numpy con las máscaras segmentadas
segmented_masks_dir = 'segmented_masks'
os.makedirs(segmented_masks_dir, exist_ok=True)

for i, mask in enumerate(segmented_masks):
     np.save(os.path.join(segmented_masks_dir, f'segmented_mask_{i}.npy'), mask)
```
